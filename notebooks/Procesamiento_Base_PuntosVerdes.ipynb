```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/Users/cristobalmorano/code/nadiasalmen/waiste/raw_data/puntos-verdes.csv", error_bad_lines=False, sep = ";")
data.head(2)
```

```python
data.drop(columns = ["id", "mas_info", "tipo", "calle", "altura", "calle2"], inplace = True)
data.head(2)
```

```python
#clean WKT column

data['WKT'] = data['WKT'].str.replace('POINT','')
data['WKT'] = data['WKT'].str.replace('(','')
data['WKT'] = data['WKT'].str.replace(')','')
data['WKT'] = data['WKT'].str.replace(' ','')
data['materiales'] = data['materiales'].str.replace('RAEES / ','')
data.head(2)
```

```python
#Split latitud and longitud from WKT column and drop

data['WKT'] = data['WKT'].str.replace('-','', 1)
data['Lat'], data['Long'] = data['WKT'].str.split('-', 1).str
data["Lat"] = "-" + data["Lat"]
data["Long"] = "-" + data["Long"]

data.drop(columns = "WKT", inplace = True)
data.head(5)
```

```python
#One hot encode materiales reciclables

data['Aceite'] = data['materiales'].apply(lambda x: int('Aceite vegetal usado (en botellas cerradas)' in x) if isinstance(x, str) else 0)
data['Pilas'] = data['materiales'].apply(lambda x: int('Pilas' in x) if isinstance(x, str) else 0)
data['Papel'] = data['materiales'].apply(lambda x: int('Papel' in x) if isinstance(x, str) else 0)
data['Cartón'] = data['materiales'].apply(lambda x: int('Cartón' in x) if isinstance(x, str) else 0)
data['Vidrio'] = data['materiales'].apply(lambda x: int('Vidrio' in x) if isinstance(x, str) else 0)
data['Metal'] = data['materiales'].apply(lambda x: int('Metal' in x) if isinstance(x, str) else 0)
data['Plástico'] = data['materiales'].apply(lambda x: int('Plástico' in x) if isinstance(x, str) else 0)
data['Tetra-brik'] = data['materiales'].apply(lambda x: int('Tetra-brik' in x) if isinstance(x, str) else 0)
data['Telgopor'] = data['materiales'].apply(lambda x: int('Telgopor' in x) if isinstance(x, str) else 0)
data['Organicos'] = data['materiales'].apply(lambda x: int('Organicos' in x) if isinstance(x, str) else 0)
data.head(10)
```
